```python
v = np.array([1,2,3])  # v has shape (3,)
w = np.array([4,5,7])    # w has shape (2,)
print("Printing v:\n",v)
print("Printing w:\n",w)
shaped=np.reshape(v,(3,1))
print("Printing shaped:\n",shaped)
print("Printing shaped2:\n",v.T)

print("printing outer product:\n",shaped*w)

print("printing outer product w/func:\n",np.outer(v,w))
print("printing outer product: w/func\n",np.outer(w,v))
```
